## Correlação entre Variaveis

```python
corr = df_encoded.corr()
```

```python
import numpy as np

fig, ax = plt.subplots(figsize=(20,16))
ax = sns.heatmap(np.round(corr, 2), vmax=1, vmin=-1, center=0,
                 square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .5})
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir a variável alvo
target_var = 'Churn_Yes'

# Definir o limiar mínimo de correlação absoluta para seleção
limiar = 0.2

# Filtrar variáveis que tenham correlação absoluta >= limiar com a variável alvo
# Isso cria uma lista com as variáveis relevantes
variaveis_relevantes = corr.index[abs(corr[target_var]) >= limiar].tolist()

# Garantir que a variável alvo está na lista (se não estiver, adiciona)
if target_var not in variaveis_relevantes:
    variaveis_relevantes.append(target_var)

# Criar uma matriz de correlação somente com as variáveis selecionadas
corr_filtrada = corr.loc[variaveis_relevantes, variaveis_relevantes]

# Gerar uma máscara para esconder o triângulo superior da matriz (incluindo diagonal)
mascara = np.triu(np.ones_like(corr_filtrada, dtype=bool))

# Plotar o heatmap com a máscara aplicada para melhor visualização
plt.figure(figsize=(12,10))
sns.heatmap(
    corr_filtrada,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.7},
    mask=mascara
)
plt.title(f'Heatmap das variáveis com correlação >= {limiar} com "{target_var}"')
plt.show()
```

## Modelos Preditivos

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
X = df_encoded.drop(columns=['Churn_Yes'])
y = df_encoded['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train_scaled, y_train)
```

### Regressão Logistica

```python
# Instanciar e treinar
lr = LogisticRegression(random_state=42)
lr.fit(X_train_bal, y_train_bal)

# Previsões
y_pred_lr = lr.predict(X_test_scaled)
y_prob_lr = lr.predict_proba(X_test_scaled)[:, 1]

# Avaliação
print("Regressão Logística")
print("Acurácia:", accuracy_score(y_test, y_pred_lr))
print("ROC AUC:", roc_auc_score(y_test, y_prob_lr))
print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

### RandomForest

```python
# Instanciar e treinar
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_bal, y_train_bal)

# Previsões
y_pred_rf = rf.predict(X_test_scaled)
y_prob_rf = rf.predict_proba(X_test_scaled)[:, 1]

# Avaliação
print("Random Forest")
print("Acurácia:", accuracy_score(y_test, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test, y_prob_rf))
print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

### Comparação


```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

# Calcular as curvas Precision-Recall
prec_log, rec_log, _ = precision_recall_curve(y_test, y_prob_lr)
prec_rf, rec_rf, _ = precision_recall_curve(y_test, y_prob_rf)

# Calcular AUC das curvas
auc_log = auc(rec_log, prec_log)
auc_rf = auc(rec_rf, prec_rf)

# Plotar o gráfico
plt.figure(figsize=(8, 6))
plt.plot(rec_log, prec_log, label=f"Regressão Logística (AUC={auc_log:.3f})")
plt.plot(rec_rf, prec_rf, label=f"Random Forest (AUC={auc_rf:.3f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall - Comparação de Modelos")
plt.legend()
plt.grid(True)
plt.show()
```

A Regressão Logística apresentou melhor capacidade de separação entre as classes (ROC AUC superior) e recall significativamente mais alto para churn, o que a torna adequada quando a estratégia de retenção é capturar o maior número possível de clientes em risco, mesmo com maior número de falsos positivos.

O Random Forest apresentou acurácia ligeiramente superior e maior precisão para churn, sendo mais indicado quando se deseja focar apenas nos clientes com alta probabilidade de cancelamento, reduzindo o custo de ações de retenção, mas aceitando perder parte dos churners.

## Relatório

O conjunto de dados, contido em dados_tratados (1).csv, possui 7.043 registros de clientes. Uma análise inicial da variável alvo Churn indicou um desbalanceamento de classes, com 26,6% dos clientes classificados como churn (evasão) e 73,4% como não-churn. Este desbalanceamento justifica a utilização de técnicas como a amostragem sintética (SMOTE) na etapa de modelagem.

O processo de preparação dos dados foi robusto e incluiu:

Limpeza e Codificação: Colunas categóricas foram transformadas em variáveis numéricas através de One-Hot Encoding com drop_first=True para evitar multicolinearidade.

Tratamento de Nulos: Onze registros com valores ausentes nas colunas Total.Day e account.Charges.Total foram removidos.

Análise de Multicolinearidade (VIF): Foi realizada uma análise criteriosa do Fator de Inflação de Variância (VIF) para identificar e remover variáveis redundantes. Variáveis com VIF infinito (como phone.PhoneService_Yes) e outras com alta correlação (como Total.Day e account.Charges.Total) foram removidas iterativamente até que os valores de VIF se mostrassem aceitáveis, garantindo a robustez estatística dos modelos.

A análise de correlação e as visualizações gráficas permitiram identificar os seguintes fatores como principais influenciadores da evasão:

Tipo de Contrato: Este é o fator mais determinante. A análise de correlação mostrou que contratos de "Um Ano" e "Dois Anos" têm uma forte correlação negativa com o churn. Em contrapartida, clientes com contratos "Mês a Mês" são significativamente mais propensos a cancelar.

Tempo de Contrato (Tenure): O gráfico de boxplot da relação entre customer.tenure e Churn_Yes deixou claro que clientes que evadem possuem um tempo de contrato consideravelmente menor do que os que permanecem. Clientes novos representam, portanto, um grupo de alto risco.

Cobranças e Tipo de Serviço:

Cobrança Mensal: Há uma correlação positiva entre account.Charges.Monthly e o churn. O boxplot da variável account.Charges.Total também indica que os clientes que cancelam tendem a ter um gasto total acumulado menor, o que está diretamente ligado ao menor tempo de contrato.

Serviço de Internet: A análise de correlação mostrou que o serviço de "Fibra Ótica" está positivamente correlacionado com o churn, provavelmente por estar associado a cobranças mensais mais altas.

Após a preparação, os dados foram divididos em conjuntos de treino (70%) e teste (30%), com estratificação pela variável alvo para manter a proporção de churn em ambas as amostras. O conjunto de treino foi padronizado (StandardScaler) e balanceado com SMOTE. Os modelos foram avaliados no conjunto de teste, que permaneceu com a distribuição original.

A Regressão Logística se destacou pelo alto Recall (81%) para a classe de churn. Isso significa que o modelo é excelente para identificar a grande maioria dos clientes que realmente irão cancelar, embora ao custo de uma precisão menor (mais falsos positivos).

O Random Forest apresentou uma acurácia geral maior e uma precisão superior (58%). Isso indica que, quando este modelo prevê um churn, a chance de estar correto é maior, mas ele falha em identificar uma parcela maior de churners em potencial.

A curva Precision-Recall (P-R) confirmou a superioridade da Regressão Logística (AUC-PR = 0,630) para o objetivo de maximizar a detecção de churners.


Com base nos fatores identificados e no desempenho dos modelos, as seguintes estratégias são recomendadas:

Foco em Contratos de Longo Prazo:

Ação: Criar campanhas agressivas com descontos ou benefícios (ex: upgrade de serviço) para clientes do plano "Mês a Mês" migrarem para contratos anuais ou bianuais.

Justificativa: O tipo de contrato é o preditor mais forte de retenção.

Programa de Onboarding e Fidelização Inicial:

Ação: Desenvolver um programa de acompanhamento para novos clientes durante os primeiros 3 a 6 meses, oferecendo suporte proativo e talvez um pequeno bônus ou desconto após o sexto mês de contrato.

Justificativa: O churn é significativamente maior entre clientes com baixo tenure.

Otimização de Pacotes e Preços:

Ação: Para clientes com serviço de fibra ótica e altas cobranças mensais, oferecer pacotes de serviços agregados (ex: TV a cabo, streaming) para aumentar a percepção de valor e diluir o custo percebido do serviço principal.

Justificativa: A combinação de fibra ótica e alta cobrança mensal é um fator de risco.

Uso Estratégico dos Modelos Preditivos:

Ação com Regressão Logística: Utilizar este modelo para gerar uma lista ampla de clientes "potencialmente em risco". Direcionar a esta lista ações de retenção de baixo custo e grande escala, como campanhas de e-mail marketing com ofertas personalizadas.

Ação com Random Forest: Utilizar este modelo para gerar uma lista menor e mais precisa de clientes com "alta probabilidade de churn". Direcionar a este grupo ações de retenção de alto valor, como uma ligação de um especialista ou ofertas de desconto mais substanciais.

A análise demonstrou que a evasão de clientes na TellecomX2 é um fenômeno previsível e influenciado por fatores claros, principalmente relacionados a contratos e custos. A implementação de estratégias de retenção direcionadas, apoiadas pelos modelos preditivos desenvolvidos, tem um alto potencial para reduzir a taxa de churn e aumentar a lucratividade da empresa.
